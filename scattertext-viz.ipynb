```python
import numpy as np
import spacy
from sklearn.linear_model import LogisticRegression

from scattertext import SampleCorpora
from scattertext import produce_scattertext_explorer
from scattertext.CorpusFromPandas import CorpusFromPandas
from scattertext.Scalers import scale
import scattertext as st
nlp = spacy.load('en')
```

```python
df = pd.read_csv('sentiment.csv')
```

```python
df['sentiment'] = ['positive' if score > 0.2 else 'negative' for score in df['sentiment']]
```

```python
df
```

```python
corpus = CorpusFromPandas(df,
                          category_col='sentiment',
                          text_col='text',
                          nlp=nlp).build()
```

```python
term_freq_df = corpus.get_term_freq_df()
```

```python
feat_builder = st.FeatsFromOnlyEmpath()
empath_corpus = st.CorpusFromParsedDocuments(df, category_col='sentiment',
                                             feats_from_spacy_doc=feat_builder, 
                                             parsed_col='text').build()
```

```python

html = st.produce_scattertext_explorer(empath_corpus,
                                        category='positive',
                                        category_name='positive',
                                        not_category_name='negative',
                                        width_in_pixels=1000,
                                        use_non_text_features=True,
                                        use_full_doc=True,
                                        topic_model_term_lists=feat_builder.get_top_model_term_lists())
open("scattertext-viz.html", 'wb').write(html.encode('utf-8'))
```
